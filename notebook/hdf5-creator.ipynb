```python
import h5py
import numpy as np
import pickle
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
def save_hdf5(embedding_path, label_path, mask_path, output_path):
    with open(embedding_path, 'rb') as f:
        embeddings = pickle.load(f)
    with open(label_path, 'rb') as f:
        labels = pickle.load(f)
    with open(mask_path, 'rb') as f:
        masks = pickle.load(f)

    with h5py.File(output_path, 'w') as hf:
        hf.create_dataset('embeddings', data=np.array(embeddings, dtype=np.float32), compression="gzip")
        hf.create_dataset('labels', data=np.array(labels, dtype=np.int64))
        hf.create_dataset('masks', data=np.array(masks, dtype=bool))
```

```python
save_hdf5('/content/drive/MyDrive/dataset/train_embeddings.pkl',
          '/content/drive/MyDrive/dataset/train_labels.pkl',
          '/content/drive/MyDrive/dataset/train_masks.pkl',
          '/content/drive/MyDrive/dataset/train_data.h5')
```

```python
save_hdf5('/content/drive/MyDrive/dataset/val_embeddings.pkl',
          '/content/drive/MyDrive/dataset/val_labels.pkl',
          '/content/drive/MyDrive/dataset/val_masks.pkl',
          '/content/drive/MyDrive/dataset/val_data.h5')
```

```python
save_hdf5('/content/drive/MyDrive/dataset/test_embeddings.pkl',
          '/content/drive/MyDrive/dataset/test_labels.pkl',
          '/content/drive/MyDrive/dataset/test_masks.pkl',
          '/content/drive/MyDrive/dataset/test_data.h5')
```
